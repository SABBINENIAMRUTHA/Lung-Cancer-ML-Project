SPLITING OF DATA INTO TRAIN AND TEST DATA

```python
x = df.drop(['LUNG_CANCER'],axis=1)
y = df['LUNG_CANCER']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.5,random_state=43)
x_train,y_train
```

MODEL BUILDING

NAIVE BAYES

```python
db = GaussianNB()
db.fit(x_train,y_train)
```

```python
predictions = db.predict(x_test)
predictions
```

```python
accuracy = accuracy_score(y_test,predictions)*100
print(accuracy)
```

```python
confusion = confusion_matrix(y_test,predictions)
classification = classification_report(y_test,predictions)
print(confusion)
print(classification)
```

```python
#Confusion Matrix
conf_matrix = confusion_matrix(y_test, predictions)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Lung Cancer', 'Lung Cancer'],
            yticklabels=['No Lung Cancer', 'Lung Cancer'])

plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

RANDOM FOREST

```python
 rfore=RandomForestClassifier()
 rfore.fit(x_train,y_train)
```

```python
def train_random_forest_model(data):
    # Initialize and train a Random Forest classifier
    random_forest_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
    random_forest_classifier.fit(x_train, y_train)

    return random_forest_classifier, x_test, y_test

def predict(model, x_test, y_test):
    # Make predictions on the test set
    y_pred = model.predict(x_test)
    # Generate classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    return y_pred
# Train the Random Forest model
model, x_test, y_test = train_random_forest_model(data)

# Predict using the trained model
y_pred = predict(model, x_test, y_test)
```

```python
print("Accuracy:",accuracy_score(y_test,y_pred)*100)
```

```python
cnf_matrix = confusion_matrix(y_test,y_pred)
plt.figure(figsize=(4,3))
p = sns.heatmap((cnf_matrix),annot=True,cmap="Purples",xticklabels=['No Lung Cancer', 'Lung Cancer'],
            yticklabels=['No Lung Cancer', 'Lung Cancer'],fmt='g')
plt.title("Confusion Matrix - RandomForest",y=1.1)
```
